# Principe numérique de la méthode tau

Le principe de la méthode tau est une extension de la méthode de Galerkin. Pour illustrer ce principe, considérons le cas où une seule condition aux limites de la forme $u(x_{CL}) = u_{CL}$ est présente, avec $x_{CL}$ représentant un point de l'espace.


## Ajout du terme tau et application de la méthode de Galerkin

En introduisant un terme tau dans l'équation de départ en raison d'une unique condition aux limites, nous obtenons :

$$ Lu(x) + \tau P(x) = f(x). $$

En appliquant la méthode de Galerkin (utilisant les mêmes notations que dans la première partie de ce projet), nous obtenons l'équation suivante pour tout $m \in \{0, 1, ..., N-1\}$ :

$$
\forall m \in \{0,1,...,N-1\}, \quad \sum_{n=0}^{N-1} u_n \langle\psi_m(x)|L|\phi_n(x)\rangle_2 + \tau \langle\psi_m(x)|P(x)\rangle_2 = \sum_{n=0}^{N-1} f_n \langle\psi_m(x)|\phi_n(x)\rangle_2.
$$

Cela nous donne un système avec $N+1$ inconnues ${u_0, ..., u_{N-1}, \tau}$ et $N$ équations.

En utilisant la liberté offerte par le paramètre $\tau$, nous pouvons ajouter une équation supplémentaire pour fermer le système. Pour cela, nous utiliserons la condition aux limites, ajoutant l'équation :

$$
\sum_{n=0}^{N-1} u_n |\phi_n(x_{CL})\rangle = u_{CL}.
$$

Ainsi, le système a désormais $N+1$ équations pour $N+1$ inconnues ${u_0, ..., u_{N-1}, \tau}$. Ce système peut être formulé de manière matricielle :


$$\begin{bmatrix}
\langle\psi_0|L|\phi_0\rangle_2 & \ldots & \langle\psi_0|L|\phi_{N-1}\rangle_2 & | & \langle\psi_0|P\rangle_2 \\
\vdots & \ddots & \vdots & | & \vdots \\
\langle\psi_{N-1}|L|\phi_0\rangle_2 & \ldots & \langle\psi_{N-1}|L|\phi_{N-1}\rangle_2 & | & \langle\psi_{N-1}|P\rangle_2 \\
\hline
\phi_{0}(x_{CL}) & \ldots & \phi_{N-1}(x_{CL}) & | & 0 \\
\end{bmatrix}
\begin{bmatrix}
u_0 \\
\vdots \\
u_{N-1} \\
\tau \\
\end{bmatrix}
= \begin{bmatrix}
\langle\psi_0|\phi_0\rangle_2 & \ldots & \langle\psi_0|\phi_{N-1}\rangle_2 & | & 0 \\
\vdots & \ddots & \vdots & | & \vdots \\
\langle\psi_{N-1}|\phi_0\rangle_2 & \ldots & \langle\psi_{N-1}|\phi_{N-1}\rangle_2 & | & 0 \\
0 & \ldots & 0 & | & 1 \\
\end{bmatrix}
\begin{bmatrix}
f_0 \\
\vdots \\
f_{N-1} \\
u_{CL} \\
\end{bmatrix}.
$$

## Choix du polynôme $P$

Comme expliqué dans la première partie de ce projet, il est numériquement avantageux de maximiser le nombre de zéros dans une matrice. Un choix efficace pour $P$ est un polynôme de la famille des $\psi_n$, par exemple :

$$ P = \psi_{N-1}(x). $$

Ainsi, avec $\psi_{n}(x)$ comme base orthonormale pour son produit scalaire, la formulation devient :


$$\begin{bmatrix}
\langle\psi_0|L|\phi_0\rangle_2 & \ldots & \langle\psi_0|L|\phi_{N-1}\rangle_2 & | & 0 \\
\vdots & \ddots & \vdots & | & \vdots \\
\langle\psi_{N-2}|L|\phi_0\rangle_2 & \ldots & \langle\psi_{N-2}|L|\phi_{N-2}\rangle_2 & | & 0\\
\langle\psi_{N-1}|L|\phi_0\rangle_2 & \ldots & \langle\psi_{N-1}|L|\phi_{N-1}\rangle_2 & | & 1\\
\hline
\phi_{0}(x_{CL}) & \ldots & \phi_{N-1}(x_{CL}) & | & 0 \\
\end{bmatrix}
\begin{bmatrix}
u_0 \\
\vdots \\
u_{N-1} \\
\tau \\
\end{bmatrix}= \begin{bmatrix}
\langle\psi_0|\phi_0\rangle_2 & \ldots & \langle\psi_0|\phi_{N-1}\rangle_2 & | & 0 \\
\vdots & \ddots & \vdots & | & \vdots \\
\langle\psi_{N-1}|\phi_0\rangle_2 & \ldots & \langle\psi_{N-1}|\phi_{N-1}\rangle_2 & | & 0 \\
\hline
0 & \ldots & 0 & | & 1 \\
\end{bmatrix}
\begin{bmatrix}
f_0 \\
\vdots \\
f_{N-1} \\
u_{CL} \\
\end{bmatrix}.
$$

En effectuant une permutation des deux dernières lignes, on obtient alors :


$$\begin{bmatrix}
\langle\psi_0|L|\phi_0\rangle_2 & \ldots & \langle\psi_0|L|\phi_{N-1}\rangle_2 & | & 0 \\
\vdots & \ddots & \vdots & | & \vdots \\
\langle\psi_{N-2}|L|\phi_0\rangle_2 & \ldots & \langle\psi_{N-2}|L|\phi_{N-2}\rangle_2 & | & 0\\
\phi_{0}(x_{CL}) & \ldots & \phi_{N-1}(x_{CL}) & | & 0 \\
\hline
\langle\psi_{N-1}|L|\phi_0\rangle_2 & \ldots & \langle\psi_{N-1}|L|\phi_{N-1}\rangle_2 & | & 1\\
\end{bmatrix}
\begin{bmatrix}
u_0 \\
\vdots \\
u_{N-1} \\
\tau \\
\end{bmatrix}= \begin{bmatrix}
\langle\psi_0|\phi_0\rangle_2 & \ldots & \langle\psi_0|\phi_{N-1}\rangle_2 & | & 0 \\
\vdots & \ddots & \vdots & | & \vdots \\
\langle\psi_{N-2}|\phi_0\rangle_2 & \ldots & \langle\psi_{N-2}|\phi_{N-1}\rangle_2 & | & 0 \\
0 & \ldots & 0 & | & 1 \\
\hline
\langle\psi_{N-1}|\phi_0\rangle_2 & \ldots & \langle\psi_{N-1}|\phi_{N-1}\rangle_2 & | & 0 \\
\end{bmatrix}
\begin{bmatrix}
f_0 \\
\vdots \\
f_{N-2} \\
f_{N-1} \\
u_{CL} \\
\end{bmatrix}.
$$

L'objectif étant de déterminer $u$, la valeur de $\tau$ ne nous intéresse pas. Nous pouvons donc éliminer son expression et l'équation qui le régit dans notre système. On obtient finalement :


$$\begin{bmatrix}
\langle\psi_0|L|\phi_0\rangle_2 & \ldots & \langle\psi_0|L|\phi_{N-1}\rangle_2  \\
\vdots & \ddots & \vdots  \\
\langle\psi_{N-2}|L|\phi_0\rangle_2 & \ldots & \langle\psi_{N-2}|L|\phi_{N-2}\rangle_2 \\
\phi_{0}(x_{CL}) & \ldots & \phi_{N-1}(x_{CL})  \\
\end{bmatrix}
\begin{bmatrix}
u_0 \\
\vdots \\
u_{N-1} \\
\end{bmatrix}= \begin{bmatrix}
\langle\psi_0|\phi_0\rangle_2 & \ldots & \langle\psi_0|\phi_{N-1}\rangle_2 & 0 \\
\vdots & \ddots & \vdots & \vdots \\
\langle\psi_{N-2}|\phi_0\rangle_2 & \ldots & \langle\psi_{N-2}|\phi_{N-1}\rangle_2 & 0 \\
0 & \ldots & 0 & 1 \\
\end{bmatrix}
\begin{bmatrix}
f_0 \\
\vdots \\
f_{N-2} \\
f_{N-1} \\
u_{CL} \\
\end{bmatrix}.
$$


## Écriture condensée du problème

On peut réécrire le problème de manière concise comme suit :

$$
D_{\tau}U=M_{\tau}F_{\tau},
$$

avec

- $D_{\tau}$ : la nouvelle matrice de dérivation contenant les conditions aux limites,
  
  $$ D_{\tau_{ij}} =
\begin{cases}
\langle\psi_i(x)|L|\phi_j(x)\rangle_2 & \text{pour } i \neq N-1 \\
\phi_{j}(x_{CL}) & \text{pour } i = N-1
\end{cases}, $$

- $U$ : le vecteur des coefficients de $u$ dans la base, $U_{i}=u_i$,

- $F_{\tau}$ : le vecteur des coefficients de $f$ dans la base et la condition aux limites,

  $$ F_{\tau_{i}} =
\begin{cases}
f_i & \text{pour } i \neq N \\
u_{CL} & \text{pour } i = N
\end{cases}, $$

- $M_{\tau_{ij}}$ : la matrice de changement de base complétée,
  $ M_{ij} =
\begin{cases}
\langle\psi_i(x)|\phi_j(x)\rangle_2 & \text{pour } i \neq N-1  \text{et } j \neq N \\
\delta_{N-1,j} & \text{pour } i = N-1 \\
\delta_{i,N} & \text{pour } j = N
\end{cases}. $

On retrouve ainsi un problème d'inversion de matrice similaire à celui de départ.





***
**Remarque : LHS, simplicité de la méthode tau en pratique**  

En termes de modification du terme de gauche (LHS), cette méthode se résume à la suppression de la dernière ligne, remplacée par la condition aux limites !
***

***
**Remarque : RHS, attention au terme ajouté** 

En revanche, pour le terme de droite (RHS), il est essentiel de le compléter. Si le LHS reste un produit d'une matrice $N\times N$ avec un vecteur de taille $N$, le RHS implique désormais le produit d'une matrice $N\times (N+1)$ avec un vecteur de taille $(N+1)$ pour inclure la valeur de la condition aux limites.
***

***
**Remarque : Permutation des lignes dans la représentation matricielle**  

L'ordre dans lequel nous avons écrit les équations dans la matrice $D_{\tau}$ est arbitraire. Il est tout à fait possible de permuter les lignes, mais il faut faire attention à également permuter les lignes de la matrice $M_{\tau}$.
***

## Application à un exemple simple : dérivée d'ordre un

Revenons à l'exemple jouet que nous avons utilisé dans la première partie de ce projet :
- $L=\frac{\partial}{\partial x}$,
- $|\phi_n(x)\rangle = T_n(x)$ avec son produit scalaire $ \langle ., . \rangle_1$,
- $|\psi_n(x)\rangle = U_n(x)$ avec son produit scalaire $ \langle ., . \rangle_2$.

Complétons-le avec la condition initiale arbitraire $u(1) = 0$ et ajoutons-lui un terme tau avec $P=U_{N-1}(x)$, menant au système :

$$ \frac{\partial u}{\partial x} + \tau U_{N-1}(x)= f(x), \quad u(0) = 0.$$

Nous avons alors :

$$
D_{\tau}U=M_{\tau}F_{\tau}
$$

avec

- $ D_{\tau_{ij}} =
\begin{cases}
\langle U_n(x)|L|T_m(x)\rangle_2 = n\delta_{n,m-1}\langle U_n(x)|\ U_m(x)\rangle_2 & \text{pour } i \neq N-1 & \text{car } \frac{\partial T_n}{\partial x} = nU_{n-1}(x)\\
T_{j}(1) = 1 & \text{pour } i = N-1 & \text{car } T_n(1) = \cos(0) =1
\end{cases} $

- $U=u_i$

- $ F_{\tau_{i}} =
\begin{cases}
f_i & \text{pour } i \neq N \\
0 & \text{pour } i = N
\end{cases} $

- $ M_{ij} =
\begin{cases}
\langle U_i(x)|\ T_j(x)\rangle_2 & \text{pour } i \neq N-1  \text{ et } j \neq N \\
\delta_{N-1,j} & \text{pour } i = N-1 \\
\delta_{i,N} & \text{pour } j = N
\end{cases} $

La ligne de code ci-dessous permet de représenter graphiquement cet exemple.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Définition des paramètres
N = 15

# Définition des polynômes de Chebysev du premier ordre (T_n)
def Tn(n, x):
    return np.cos(n * np.arccos(x))

# Définition des polynômes de Chebysev du second ordre (U_n)
def Un(n, x):
    numerator = np.sin((n + 1) * np.arccos(x))
    denominator = np.sin(np.arccos(x))
    return numerator / denominator

# Définition des produits scalaires
def inner_product_0(f, g):
    integrand = lambda x: f(x) * g(x) / np.sqrt(1 - x**2)
    result, _ = quad(integrand, -1, 1)
    return result

def inner_product_2(f, g):
    integrand = lambda x: f(x) * g(x) * np.sqrt(1 - x**2)
    result, _ = quad(integrand, -1, 1)
    return result

# Calcul des dérivées des polynômes de Chebysev du premier ordre
def dTn_dx(n, x):
    return n * np.sin(n * np.arccos(x)) / np.sqrt(1 - x**2)

# Calcul des coefficents de la décomposition
Dtau = np.zeros((N, N))
Mtau = np.zeros((N, N+1))


for i in range(N-1): # N-1 car avec la méthode tau remplace l'équation sur le dernier coefficient par les conditions aux limites
    for j in range(N):
        f_Ti = lambda x: Tn(i, x)
        f_Tj = lambda x: Tn(j, x)
        f_Ui = lambda x: Un(i, x)
        f_Uj = lambda x: Un(j, x)
        f_dT = lambda x: dTn_dx(j, x)

        Dtau[i, j] = inner_product_2(f_Ui, f_dT)
        Mtau[i, j] = inner_product_2(f_Ui, f_Tj)

# Ajout de la condition initiale par la méthode tau
for i in range(N):
    Dtau[N-1, i] = 1
Mtau[N-1, N] = 1

# Fonction pour afficher la matrice avec des carrés noires pour les valeurs > epsilon et blanc sinon
def plot_matrix(ax, matrix, title, epsilon=1e-5):
    ax.imshow(np.abs(matrix) > epsilon, cmap='binary', interpolation='none')

    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            ax.text(j, i, f'{matrix[i, j]:.1f}', color='red',
                     ha='center', va='center', fontsize=8)

    ax.set_title(title)

# Création de sous-graphiques
fig, axs = plt.subplots(1, 2, figsize=(12, 12))


# Affichage des matrices
plot_matrix(axs[0], Dtau, r'$D_{\tau_{ij}}$')
plot_matrix(axs[1], Mtau, r'$M_{\tau_{ij}}$')
```

Par cet exemple, il devient évident que la méthode tau se révèle efficace pour l'implémentation des conditions aux limites. Néanmoins, elle introduit un nouveau défi : la matrice inverse n'est plus diagonale ni en bande.

> Une solution astucieuse pour restaurer une matrice diagonale ou en bande consiste à apporter de légères modifications à la base initiale. Cette approche permet de conserver la sparsité de la matrice de dérivation initiale tout en rendant également creuse la contribution due aux conditions aux limites. Nous appliquerons l'astuce de recombinaison de Dirichlet pour résoudre ce problème.

# Méthode de recombinaison de Dirichlet 

## Conditions de Dirichlet et de Neumann

Les conditions de Dirichlet et de Neumann sont des types de conditions aux limites fréquemment employées dans la résolution d'équations aux dérivées partielles, en particulier dans le cadre de la physique. Voici une synthèse de chacune de ces conditions :

1. **Condition de Dirichlet :**
   - La condition de Dirichlet spécifie la valeur d'une fonction à la frontière d'un domaine.
   - Un exemple concret serait, dans le contexte de l'équation modélisant la distribution thermique dans un objet, la fixation de la température à la surface de cet objet.

2. **Condition de Neumann :**
   - La condition de Neumann spécifie le taux de variation d'une fonction à la frontière d'un domaine.
   - Plutôt que de fixer la valeur de la fonction à la frontière, la condition de Neumann fixe la dérivée normale de la fonction à cet endroit.
   - Par exemple, dans le contexte de la conduction thermique, la condition de Neumann pourrait définir le flux de chaleur à travers la frontière d'un objet.


## Préconditionnement de Dirichlet

On définit une nouvelle base d'essaie (noté avant $\phi_n(x)$): 

$$
D_n(x) = 
\begin{cases}
T_n(x), & \text{pour } n = 0,1 \\
T_n(x) - T_{n-2}(x), & \text{pour } n > 1
\end{cases}.
$$

On a alors pour cette famille au niveau des bords du type $x=1$ ou $x=-1$ :

$$
D_n(\pm 1) = 
\begin{cases}
(\pm 1)^n, & \text{pour } n = 0,1 \\
0, & \text{pour } n > 1
\end{cases}.
$$

Dans cette base, les conditions aux limites de Dirichlet ne concernent que les deux premiers coefficients du développement. Cette technique est connue sous le nom de "préconditionnement de Dirichlet" ou "recombinaison de base".

Ainsi, si on veut retrouver une matrice diagonale ou en bande, il est important de permuter les lignes de sorte que les deux coefficients soient en haut à gauche, et non pas en haut à droite. 

Visuellement on a initalement :
$$
\begin{bmatrix}
\langle\psi_0|L|D_0\rangle_2 & \ldots & \langle\psi_0|L|D_{N-1}\rangle_2  \\
\vdots & \ddots & \vdots  \\
\langle\psi_{N-2}|L|D_0\rangle_2 & \ldots & \langle\psi_{N-2}|L|D_{N-2}\rangle_2 \\
D_{0}(x_{CL}) & \ldots & D_{N-1}(x_{CL})  \\
\end{bmatrix}
\begin{bmatrix}
u_0 \\
\vdots \\
u_{N-1} \\
\end{bmatrix}= \begin{bmatrix}
\langle\psi_0|D_0\rangle_2 & \ldots & \langle\psi_0|D_{N-1}\rangle_2 & 0 \\
\vdots & \ddots & \vdots & \vdots \\
\langle\psi_{N-2}|D_0\rangle_2 & \ldots & \langle\psi_{N-2}|D_{N-1}\rangle_2 & 0 \\
0 & \ldots & 0 & 1 \\
\end{bmatrix}
\begin{bmatrix}
f_0 \\
\vdots \\
f_{N-2} \\
f_{N-1} \\
u_{CL} \\
\end{bmatrix}.
$$

On évalue $D_n$ en $x=\pm 1$ :

$$
\begin{bmatrix}
\langle\psi_0|L|D_0\rangle_2 & & \ldots & & \langle\psi_0|L|D_{N-1}\rangle_2  \\
\vdots & & \ddots & & \vdots  \\
\langle\psi_{N-2}|L|D_0\rangle_2 & &  \ldots & & \langle\psi_{N-2}|L|D_{N-2}\rangle_2 \\
(\pm 1)^n & (\pm 1)^n & 0 & \ldots & 0  \\
\end{bmatrix}
\begin{bmatrix}
u_0 \\
\vdots \\
u_{N-1} \\
\end{bmatrix}= \begin{bmatrix}
\langle\psi_0|D_0\rangle_2 & \ldots & \langle\psi_0|D_{N-1}\rangle_2 & 0 \\
\vdots & \ddots & \vdots & \vdots \\
\langle\psi_{N-2}|D_0\rangle_2 & \ldots & \langle\psi_{N-2}|D_{N-1}\rangle_2 & 0 \\
0 & \ldots & 0 & 1 \\
\end{bmatrix}
\begin{bmatrix}
f_0 \\
\vdots \\
f_{N-2} \\
f_{N-1} \\
u_{CL} \\
\end{bmatrix}.
$$

Puis on permute les lignes dans l'objectif d'avoir une matrice en bande (après calcul des produiuts scalaires).

$$
\begin{bmatrix}
(\pm 1)^n & (\pm 1)^n & 0 & \ldots & 0  \\
\langle\psi_0|L|D_0\rangle_2 & & \ldots & & \langle\psi_0|L|D_{N-1}\rangle_2  \\
\vdots & & \ddots & & \vdots  \\
\langle\psi_{N-2}|L|D_0\rangle_2 & &  \ldots & & \langle\psi_{N-2}|L|D_{N-2}\rangle_2 \\
\end{bmatrix}
\begin{bmatrix}
u_0 \\
\vdots \\
u_{N-1} \\
\end{bmatrix}= \begin{bmatrix}
0 & \ldots & 0 & 1 \\
\langle\psi_0|D_0\rangle_2 & \ldots & \langle\psi_0|D_{N-1}\rangle_2 & 0 \\
\vdots & \ddots & \vdots & \vdots \\
\langle\psi_{N-2}|D_0\rangle_2 & \ldots & \langle\psi_{N-2}|D_{N-1}\rangle_2 & 0 \\
\end{bmatrix}
\begin{bmatrix}
f_0 \\
\vdots \\
f_{N-2} \\
f_{N-1} \\
u_{CL} \\
\end{bmatrix},
$$

***
**Remarque : Diagonalité de la matrice de changement de base**

Il est avantageux que la matrice de changement de base complétée contienne un maximum de zéros pour faciliter le produit matriciel. Cependant, l'aspect crucial réside dans le fait que la matrice de dérivation avec les conditions limites soit aussi proche que possible d'une matrice diagonale. En effet, le calcul de l'inversion de matrice est bien plus coûteux que celui du produit matriciel. **On s'intéressera donc principalement dans la suite à l'optimisation de la matrice $D_{\tau}$.**
***
***
**Remarque : Expression des conditions aux limites**

Étant donné que les valeurs pratiques de la famille de polynômes $D_n$ correspondent à des valeurs de $x$ égales à 1 ou -1, il est judicieux d'effectuer un adimensionnement approprié afin d'obtenir une expression des conditions aux limites sous la forme $\pm 1$.

***

## Retour sur l'exemple de la dérivée d'ordre un

On reprend notre exemple en changeant $|\phi_n(x)\rangle$ :

$$|\phi_n(x)\rangle = D_n(x).$$


Nous avons alors :

$$
D_{\tau}U=M_{\tau}F_{\tau},
$$

avec

- $ D_{\tau_{ij}} =
\begin{cases}
\langle U_n(x)|L|D_m(x)\rangle_2 & \text{pour } i \neq 0 \\
D_{j}(1) = 1 & \text{pour } i = 0 \text{ et } j = 0 \text{ ou } 1\\
D_{j}(1) = 0 & \text{pour } i = 0 \text{ et } j >1
\end{cases} .$


La ligne de code ci-dessous permet de représenter graphiquement cet exemple.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Définition des paramètres
N = 15

# Définition des polynômes de Chebysev du premier ordre (T_n)
def Tn(n, x):
    return np.cos(n * np.arccos(x))

# Définition des polynômes de Chebysev du second ordre (U_n)
def Un(n, x):
    numerator = np.sin((n + 1) * np.arccos(x))
    denominator = np.sin(np.arccos(x))
    return numerator / denominator

# Définition des produits scalaires
def inner_product_0(f, g):
    integrand = lambda x: f(x) * g(x) / np.sqrt(1 - x**2)
    result, _ = quad(integrand, -1, 1)
    return result

def inner_product_2(f, g):
    integrand = lambda x: f(x) * g(x) * np.sqrt(1 - x**2)
    result, _ = quad(integrand, -1, 1)
    return result

# Calcul des dérivées des polynômes de Chebysev du premier ordre
def dTn_dx(n, x):
    return n * np.sin(n * np.arccos(x)) / np.sqrt(1 - x**2)

# Calcul des polynomes de Dirichlet
def Dn(n, x):
    if n == 0 or n == 1:
        return Tn(n, x)
    elif n > 1:
        return Tn(n, x) - Tn(n-2, x)
        
# Et de leurs dérivés 
def dDn_dx(n, x):
    if n == 0 or n == 1:
        return dTn_dx(n, x)
    elif n > 1:
        return dTn_dx(n, x) - dTn_dx(n-2, x)

# Calcul des coefficents de la décomposition
Dtau = np.zeros((N, N))
Mtau = np.zeros((N, N+1))


for i in range(N-1):
    for j in range(N):
        f_Ui = lambda x: Un(i, x)
        f_dD = lambda x: dDn_dx(j, x)

        Dtau[i+1, j] = inner_product_2(f_Ui, f_dD) # +1 dû à la méthode tau qui décale et supprime une ligne 

# Ajout de la condition initiale par la méthode tau
for i in range(2):
    Dtau[0, i] = 1

# Fonction pour afficher la matrice avec des carrés noires pour les valeurs > epsilon et blanc sinon
def plot_matrix(ax, matrix, title, epsilon=1e-5):
    ax.imshow(np.abs(matrix) > epsilon, cmap='binary', interpolation='none')

    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            ax.text(j, i, f'{matrix[i, j]:.1f}', color='red',
                     ha='center', va='center', fontsize=8)

    ax.set_title(title)

# Création de sous-graphiques
fig = plt.figure(figsize=(8, 8))
axs = fig.add_subplot(1, 1, 1)

# Affichage des matrices
plot_matrix(axs, Dtau, r'$D_{\tau_{ij}}$')
```

On observe que la méthode permet d'éliminer la ligne du bas, rétablissant ainsi une structure de matrice à bande.

***
**Remarque : Matrice à bande**

Il est essentiel de noter que l'obtention de matrices diagonales n'est réalisable que dans des cas très simples. Dans la suite, nous nous contenterons d'obtenir des matrices à bandes / matrices bandes. Cela signifie qu'il s'agit de matrices creuses, dont les coefficients non nuls sont limités à une bande diagonale, englobant la diagonale principale et éventuellement une ou plusieurs diagonales de chaque côté. Ces matrices, après les matrices diagonales, sont les matrices les plus simples à inverser et à stocker.

# Systèmes d'équations au-delà du premier ordre

L'objectif de cette section est de se familiariser avec le traitement des équations comportant des dérivées d'ordre supérieur à un.

## Introduction : changement de point de vue 

### Principe : d'une équation d'ordre $N$ à $N$ équations d'ordre un

Depuis le début, nous avons adopté une approche matricielle. Il est donc naturel de considérer la transformation d'une équation d'ordre N en un système d'équations d'ordre un, une pratique courante en mathématiques. C'est précisément ce que nous allons entreprendre ici. Notons d'ailleurs que le nombre de termes tau à ajouter est directement lié au nombre de dérivées présentes dans notre problème. Ainsi, gérer les termes tau pour une équation d'ordre un est plus aisé, il suffira d'ajouter un terme tau à chaque équation d'ordre un.

### Exemple : l'équation de Poisson 
Considérons par exemple l'équation de Poisson à une dimension :

$$ \frac{\partial^2 u}{\partial x^2} = f(x), \quad u(-1) = 0, \quad \frac{\partial u}{\partial x}(1) = 0. $$

On a ici une équation d'ordre deux. Il est facile de l'écrire comme deux équations d'ordre un, comme suit :

$$
\begin{cases}
\frac{\partial u}{\partial x} = u_x \\
\frac{\partial u_x}{\partial x} = f(x) \\
\end{cases}, \quad u(-1) = 0, \quad u_x(1) = 0. $$

Nous pouvons écrire cette équation matriciellement, comme ceci :



$$\begin{bmatrix}
\frac{\partial}{\partial x} & -1 \\
0 & \frac{\partial}{\partial x}
\end{bmatrix}
\begin{bmatrix}
u \\
u_x
\end{bmatrix}
=\begin{bmatrix}
0 \\
f(x)
\end{bmatrix}.
$$

En simplifiant, on obtient l'équation suivante :
$$L \cdot X = F,$$
où :
- $X$ est le vecteur d'état des variables, $X=\begin{bmatrix} u \\ u_x \end{bmatrix},$
- $L$ représente la matrice des opérateurs, définie comme $L = \begin{bmatrix} \frac{\partial}{\partial x} & -1 \\ 0 & \frac{\partial}{\partial x} \end{bmatrix},$
- $F$ est le vecteur des seconds membres, $F=\begin{bmatrix} 0 \\ f(x) \end{bmatrix}.$

### Généralisation à tout système 

De manière général, pour un système d'ordre $S$ avec $S$ conditions aux limites, on obtiendra un système d'équations pour les variables de la forme :

$$
\forall i \in \{1,...,S\}, \quad \sum_{j=1}^{S} L_{i,j}u_j(x)=f_i(x), \quad u_i(x_{CL, i}) = u_{CL, i}.
$$

Que l'on peut récrire :
$$L \cdot X = F,$$
où :
- $X$ est le vecteur d'état des variables, $X=\begin{bmatrix} u_1 \\ u_2 \\ \vdots \\ u_S \end{bmatrix}$
- $L$ est la matrice des opérateurs, $L = \begin{bmatrix} L_{1,0} & L_{1,1} & \cdots & L_{1,S} \\ L_{2,0} & L_{2,1} & \cdots & L_{2,S} \\ \vdots & \vdots & \ddots & \vdots \\ L_{S,0} & L_{S,1} & \cdots & L_{S,S} \end{bmatrix}$
- $F$ est le vecteur des seconds membres, $F=\begin{bmatrix} f_1(x) \\ f_2(x) \\ \vdots \\ f_S(x) \end{bmatrix}$


## Résolution des systèmes sous forme d'équations du premier ordre

### Principe de base : une méthode en six étapes

Pour résoudre les systèmes sous forme d'équations du premier ordre, nous suivrons le même principe qu'auparavant. 
1. Décomposition sous forme d'un système d'ordre 1
2. Ajout des termes tau et choix des polynômes $P$
3. Décomposition dans la base d'essais
4. Projection par la base de test
5. Écriture matricielle
6. Permutation des lignes et suppression des termes tau


### Exemple : Retour sur l'équation de Poisson

Reprenons l'exemple de l'équation de Poisson :

$$ \frac{\partial^2 u}{\partial x^2} = f(x), \quad u(-1) = 0, \quad \frac{\partial u}{\partial x}(1) = 0. $$

Pour simplifier, on prendra $f(x)=0$, cela ne change en rien l'esprit de la méthode. Cela simplifie seulement les calculs.

1. Décomposition sous formes d'un système d'ordre 1 :
   
$$
\begin{cases}
\frac{\partial u}{\partial x} - u_x = 0 \\
\frac{\partial u_x}{\partial x} = 0 \\
\end{cases}, \quad u(-1) = 0, \quad u_x(1) = 0. $$

2. Ajout des termes tau et choix des polynômes $P$, on prend pour l'exemple $P_1(x)=P_2(x)=U_{N-1}(x)$ :

$$
\begin{cases}
\frac{\partial u}{\partial x} - u_x + \tau_1 U_{N-1}(x)= 0 \\
\frac{\partial u_x}{\partial x} + \tau_2 U_{N-1}(x) = 0 \\
\end{cases}, \quad u(-1) = 0, \quad u_x(1) = 0. $$

3. Décomposition dans la base d'essaies $D_n$, on note $L=\frac{\partial}{\partial x}$ :

$$
\begin{cases}
\sum_{n=0}^{N-1}u_nL|D_n(x)\rangle - (u_x)_n|D_n(x)\rangle + \tau_1 U_{N-1}(x)= 0 \\
\sum_{n=0}^{N-1}(u_x)_nL|D_n(x)\rangle + \tau_2 U_{N-1}(x) = 0 \\
\end{cases}, \quad u(-1) = 0, \quad u_x(1) = 0. $$

4. Projection par la base de test $\langle U_m(x)|$ :

$$
\begin{cases}
\sum_{n=0}^{N-1}u_n\langle U_m(x)|L|D_n(x)\rangle_2 - (u_x)_n\langle U_m(x)|D_n(x)\rangle_2 + \tau_1 \langle U_m(x)|U_{N-1}(x)\rangle_2= 0 \\
\sum_{n=0}^{N-1}(u_x)_n\langle U_m(x)|L|D_n(x)\rangle_2 + \tau_2\langle U_m(x)| U_{N-1}(x)\rangle_2= 0 \\
\end{cases}, \quad u(-1) = 0, \quad u_x(1) = 0. $$

5. Ecriture matricielle :

$$\begin{bmatrix}
\langle U_0|L|D_0\rangle_2 & \ldots & \langle U_0|L|D_{N-1}\rangle_2 & | & 0  & | & -\langle U_0|D_0\rangle_2 & \ldots & -\langle U_0|D_{N-1}\rangle_2 & | & 0 \\
\vdots & \ddots & \vdots & | & \vdots & | & \vdots & \ddots & \vdots & | & \vdots \\
\langle U_{N-2}|L|D_0\rangle_2 & \ldots & \langle U_{N-2}|L|D_{N-2}\rangle_2 & | & 0  & | & -\langle U_{N-2}|D_0\rangle_2 & \ldots & -\langle U_{N-2}|D_{N-2}\rangle_2 & | & 0\\
\langle U_{N-1}|L|D_0\rangle_2 & \ldots & \langle U_{N-1}|L|D_{N-1}\rangle_2 & | & 1  & | & -\langle U_{N-1}|D_0\rangle_2 & \ldots & -\langle U_{N-1}|D_{N-1}\rangle_2 & | & 0\\
\hline
0 & \ldots & 0 & | & 0  & | & \langle U_0|L|D_0\rangle_2 & \ldots & \langle U_0|L|D_{N-1}\rangle_2 & | & 0 \\
\vdots & \ddots & \vdots & | & \vdots & | & \vdots & \ddots & \vdots & | & \vdots \\
0 & \ldots & 0 & | & 0  & | & \langle U_{N-2}|L|D_0\rangle_2 & \ldots & \langle U_{N-2}|L|D_{N-1}\rangle_2 & | & 0 \\
0 & \ldots & 0 & | & 0  & | & \langle U_{N-1}|L|D_0\rangle_2 & \ldots & \langle U_{N-1}|L|D_{N-1}\rangle_2 & | & 1 \\
\hline
0  & \ldots & 0 & | & 0 & | & D_{0}(1) & \ldots & D_{N-1}(1) & | & 0 \\
D_{0}(-1) & \ldots & D_{N-1}(-1) & | & 0 & | & 0 & \ldots & 0 & | & 0 \\
\end{bmatrix}
\begin{bmatrix}
u_0 \\
\vdots \\
u_{N-1} \\
\tau_1 \\
(u_x)_0 \\
\vdots \\
(u_x)_{N-1} \\
\tau_2 \\
\end{bmatrix}= 0.$$

6. Permutation des lignes et suppression des termes tau :

$$\begin{bmatrix}
D_{0}(-1) & \ldots & D_{N-1}(-1) & | & 0 &\ldots & 0 &  \\
\langle U_0|L|D_0\rangle_2 & \ldots & \langle U_0|L|D_{N-1}\rangle_2  & | & -\langle U_0|D_0\rangle_2 & \ldots & -\langle U_0|D_{N-1}\rangle_2 &  \\
\vdots & \ddots & \vdots & | & \vdots & \ddots & \vdots \\
\langle U_{N-2}|L|D_0\rangle_2 & \ldots & \langle U_{N-2}|L|D_{N-2}\rangle_2  & | & -\langle U_{N-2}|D_0\rangle_2 & \ldots & -\langle U_{N-2}|D_{N-2}\rangle_2 & \\
\hline
0  & \ldots & 0 & | & D_{0}(1) & \ldots & D_{N-1}(1) \\
0 & \ldots & 0 & |  & \langle U_0|L|D_0\rangle_2 & \ldots & \langle U_0|L|D_{N-1}\rangle_2  \\
\vdots & \ddots  & \vdots & | & \vdots & \ddots  & \vdots \\
0 & \ldots & 0   & | & \langle U_{N-2}|L|D_0\rangle_2 & \ldots & \langle U_{N-2}|L|D_{N-1}\rangle_2   \\
\end{bmatrix}
\begin{bmatrix}
u_0 \\
\vdots \\
u_{N-1} \\
(u_x)_0 \\
\vdots \\
(u_x)_{N-1} \\
\end{bmatrix}= 0.$$

On observe l'apparition d'une décomposition bloc par bloc. On écrira ce système (noté $b$ pour bloc) de la manière suivante : 
$$
R X = 0.
$$

Le code ci-dessous permet de visualiser ce résultat. Pour une représentation matricielle plus claire, une case bleue sera utilisée pour une valeur négative, une case rouge pour une valeur positive, et une case blanche pour une valeur nulle.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Définition des paramètres
N = 15

# Définition des polynômes de Chebysev du premier ordre (T_n)
def Tn(n, x):
    return np.cos(n * np.arccos(x))

# Définition des polynômes de Chebysev du second ordre (U_n)
def Un(n, x):
    numerator = np.sin((n + 1) * np.arccos(x))
    denominator = np.sin(np.arccos(x))
    return numerator / denominator

# Définition des produits scalaires
def inner_product_0(f, g):
    integrand = lambda x: f(x) * g(x) / np.sqrt(1 - x**2)
    result, _ = quad(integrand, -1, 1)
    return result

def inner_product_2(f, g):
    integrand = lambda x: f(x) * g(x) * np.sqrt(1 - x**2)
    result, _ = quad(integrand, -1, 1)
    return result

# Calcul des dérivées des polynômes de Chebysev du premier ordre
def dTn_dx(n, x):
    return n * np.sin(n * np.arccos(x)) / np.sqrt(1 - x**2)

# Calcul des polynomes de Dirichlet
def Dn(n, x):
    if n == 0 or n == 1:
        return Tn(n, x)
    elif n > 1:
        return Tn(n, x) - Tn(n-2, x)
        
# Et de leurs dérivés 
def dDn_dx(n, x):
    if n == 0 or n == 1:
        return dTn_dx(n, x)
    elif n > 1:
        return dTn_dx(n, x) - dTn_dx(n-2, x)

# Calcul des coefficents de la décomposition
Dtau = np.zeros((2*N, 2*N))


for i in range(N-1):
    for j in range(N):
        f_Ui = lambda x: Un(i, x)
        f_Dj = lambda x: Dn(j, x)
        f_dD = lambda x: dDn_dx(j, x)

        Lij = inner_product_2(f_Ui, f_dD)
        Dtau[i+1, j] = Lij # +1 dû à la méthode tau qui décale et supprime une ligne 
        Dtau[N+i+1, N+j] = Lij # +1 dû à la méthode tau qui décale et supprime une ligne 

        Dtau[i+1, N+j] = -inner_product_2(f_Ui, f_Dj) # +1 dû à la méthode tau qui décale et supprime une ligne 
        
# Ajout de la condition initiale par la méthode tau
#D_n(1) = 1 pour n = 0 ou 1
#D_n(-1) = 1 pour n = 0 et - 1 pour n = 1
Dtau[0, 0] = 1 
Dtau[0, 1] = -1 
Dtau[N,N] = 1 
Dtau[N,N+ 1] = 1 

def plot_L(ax, matrix, title, epsilon=1e-5):
    colors = np.zeros_like(matrix, dtype=np.dtype((float, 4)))  # RGBA

    for i in range(2*N):
        for j in range(2*N):
            value = matrix[i, j]
            if abs(value) < epsilon:
                colors[i, j] = [1, 1, 1, 1]  # Blanc pour les valeurs proches de zéro
            elif value > 0:
                colors[i, j] = [0, 0, 1, 1]  # Bleu pour les valeurs positives
            else:
                colors[i, j] = [1, 0, 0, 1]  # Rouge pour les valeurs négatives

    ax.imshow(colors, interpolation='none')
    # Contour noir
    ax.contour(colors[:, :, 0], levels=[2], colors='black', linewidths=2)
    
    # Grille
    ax.set_xticks(np.arange(0, 2*N, 1))
    ax.set_yticks(np.arange(0, 2*N, 1))
    ax.set_xticks(np.arange(-0.5, 2*N, 1), minor=True)
    ax.set_yticks(np.arange(-0.5, 2*N, 1), minor=True)
    ax.grid(which="minor", color="black", linestyle='-', linewidth=0.5)
    ax.tick_params(which="major", size=0)  # Désactiver les ticks majeurs

    
    ax.set_xticklabels([])  # Désactiver les labels des axes x
    ax.set_yticklabels([])  # Désactiver les labels des axes y

    ax.set_title(title)

# Création de sous-graphiques
fig = plt.figure(figsize=(8, 8))
axs = fig.add_subplot(1, 1, 1)

# Affichage des matrices
plot_L(axs, Dtau, r'$R_{i,j}$')
```

### Généralisation : Application de la méthode bloc par bloc

En conclusion, l'application de la méthode tau à la résolution des systèmes du type :

$$
\forall i \in \{1,...,S\}, \quad \sum_{j=1}^{S} L^{i,j}u_j(x)=f_i(x), \quad u_i(x_{CL, i}) = u_{CL, i},
$$

se résume simplement à l'application de la méthode à chaque bloc $L^{i,j}$ de la décomposition. Cela se traduit par un travail bloc par bloc, **les blocs résultants seront notés $R^{i,j}$ et la matrice globale $R$**.

La méthode est efficace. Cependant, on constate une nouvelle fois la perte de la structure de la matrice à bande. Comment peut-on retrouver cette caractéristique matricielle ?

> La solution réside dans une astuce utilisant les matrices élémentaires, il s'agit d'effectuer un renversement de produit de Kronecker.

## Renversement de produit de Kronecker

L'idée principale de cette méthode repose sur la possibilité d'écrire le résultat sous forme d'un produit de Kronecker entre la matrice $L'$, définie par les éléments $L'_{i,j}$, et des matrices élémentaires.

### Définitions : Produit de Kronecker et matrices élementaires 

#### Produit de Kronecker

Le **produit de Kronecker** entre deux matrices $A$ de taille $m \times n$ et $B$ de taille $p \times q$ est noté $A \otimes B$. La résultante est une matrice de taille $mp \times nq$, définie comme suit :

$$
A \otimes B = \begin{bmatrix} a_{1,1}B & \ldots & a_{1,n}B \\ \vdots & \ddots & \vdots \\ a_{m,1}B & \ldots & a_{m,n}B \end{bmatrix},
$$

où $a_{i,j}B$ représente la multiplication de chaque élément de la matrice $A$ par la matrice $B$.

#### Commutativité du Produit de Kronecker 

Cette propiété est tiré du papier "On the Kronecker Product" de Kathrin Schäcke".

Pour $A \in \mathbb{M}_{p,q}$ et $B \in \mathbb{M}_{r,s}$, on a :

$$B \otimes A = S_{p,r}(A \otimes B)S_{q,s}^T,$$ 

où

$$S_{m,n} = \sum_{i=1}^{m} (e_i^T \otimes I_n \otimes e_i) = \sum_{i=1}^{n}(e_j \otimes I_m \otimes e_j^T)$$

est la **matrice de permutation parfaite de shuffle** (perfect shuffle permutation matrix). Elle est inversible et on a $S_{m,n}^{-1}=S_{m,n}^T$. Elle est décrite en détail dans "Kronecker Products and Matrix Calculus With Applications" de A. Graham. 

#### Matrice élémentaire $E^{i,j}$

La matrice élémentaire $E^{i,j}$ est une matrice avec des éléments nuls, sauf à la position $(i, j)$ où elle contient la valeur $1$ (*c'est-à-dire* $E^{i,j}_{k,l}=\delta_{ik}\delta_{jl}$). Elle est représentée comme suit :

$$
E^{i,j} = \begin{bmatrix}
    0 & \ldots & 0 & 0 \\
    \vdots & \ddots & \vdots & 0 \\
    0 & \ldots & 1 & 0 \\
    0 & \ldots & 0 & 0 \\
\end{bmatrix}.
$$


### Rédaction alternative de la méthode bloc par bloc

Avec ces définitions, on peut écrire :
$$
R = \sum_{i,j=1}^{S} E^{i,j} \otimes R^{i,j}.
$$

On définit $\tilde{R}$ par : 
$$
\tilde{R} = \sum_{i,j=1}^{S}  R^{i,j} \otimes E^{i,j}.
$$

Une propriété importante que l'on va utiliser ici stipule que **si $R$ a des blocs à bande, alors $\tilde{R}$ est globalement à bande**, et vice versa. Ainsi, avec la propriété sur le renversement, sachant que $R \in \mathbb{M}_{NS,NS}$ et $R^{i,j}\in \mathbb{M}_{N,N}$, on a :
$$\tilde{R} = \sum_{i,j=1}^{S}  R^{i,j} \otimes E^{i,j} = \sum_{i,j=1}^{S} S_{N,SN}(E^{i,j} \otimes R^{i,j})S_{N,SN}^T = S_{N,SN}(\sum_{i,j=1}^{S} E^{i,j} \otimes R^{i,j})S_{N,SN}^T = S_{N,SN}RS_{N,SN}^T 
\Leftrightarrow R = S_{N,SN}^T\tilde{R}S_{N,SN}.$$
On peut donc choisir de travailler, soit avec $R$ soit avec $\tilde{R}$. On choisira la matrice la plus simple à inverser. Cette méthode est appelée **renversement du produit de Kronecker**.



### Application au résultat de l'équation de Poisson

Reprenons l'exemple de l'équation de Poisson :

$$ \frac{\partial^2 u}{\partial x^2} = f(x), \quad u(-1) = 0, \quad \frac{\partial u}{\partial x}(1) = 0. $$

La matrice $R$ que l'on a obtenu est sous forme de bloc à bande, essayons donc d'appliquer le renversement du produit de Kronecker.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Définition des paramètres
N = 15

# Définition des polynômes de Chebysev du premier ordre (T_n)
def Tn(n, x):
    return np.cos(n * np.arccos(x))

# Définition des polynômes de Chebysev du second ordre (U_n)
def Un(n, x):
    numerator = np.sin((n + 1) * np.arccos(x))
    denominator = np.sin(np.arccos(x))
    return numerator / denominator

# Définition des produits scalaires
def inner_product_0(f, g):
    integrand = lambda x: f(x) * g(x) / np.sqrt(1 - x**2)
    result, _ = quad(integrand, -1, 1)
    return result

def inner_product_2(f, g):
    integrand = lambda x: f(x) * g(x) * np.sqrt(1 - x**2)
    result, _ = quad(integrand, -1, 1)
    return result

# Calcul des dérivées des polynômes de Chebysev du premier ordre
def dTn_dx(n, x):
    return n * np.sin(n * np.arccos(x)) / np.sqrt(1 - x**2)

# Calcul des polynomes de Dirichlet
def Dn(n, x):
    if n == 0 or n == 1:
        return Tn(n, x)
    elif n > 1:
        return Tn(n, x) - Tn(n-2, x)
        
# Et de leurs dérivés 
def dDn_dx(n, x):
    if n == 0 or n == 1:
        return dTn_dx(n, x)
    elif n > 1:
        return dTn_dx(n, x) - dTn_dx(n-2, x)

# Calcul des coefficents de la décomposition
Dtau = np.zeros((2*N, 2*N))

for i in range(N-1):
    for j in range(N):
        f_Ui = lambda x: Un(i, x)
        f_Dj = lambda x: Dn(j, x)
        f_dD = lambda x: dDn_dx(j, x)

        Lij = inner_product_2(f_Ui, f_dD)
        Dtau[i+1, j] = Lij # +1 dû à la méthode tau qui décale et supprime une ligne 
        Dtau[N+i+1, N+j] = Lij # +1 dû à la méthode tau qui décale et supprime une ligne 

        Dtau[i+1, N+j] = -inner_product_2(f_Ui, f_Dj) # +1 dû à la méthode tau qui décale et supprime une ligne 
        
# Ajout de la condition initiale par la méthode tau
#D_n(1) = 1 pour n = 0 ou 1
#D_n(-1) = 1 pour n = 0 et - 1 pour n = 1
Dtau[0, 0] = 1 
Dtau[0, 1] = -1 
Dtau[N,N] = 1 
Dtau[N,N+ 1] = 1 

### Calcul du renveserment de Kronecker 

# Dtau sous forme de bloc
D11 = Dtau[:N,:N]
D12 = Dtau[:N,N:]
D21 = Dtau[N:,:N]
D22 = Dtau[N:,N:]

# matrice élemntaire

def matrice_elementaire(i, j, taille):
    E = np.zeros((taille, taille))
    E[i, j] = 1
    return E
    
E11 = matrice_elementaire(0, 0, 2)
E12 = matrice_elementaire(0, 1, 2)
E21 = matrice_elementaire(1, 0, 2)
E22 = matrice_elementaire(1, 1, 2)

# calcul de DtauTilde
DtauTilde = np.zeros((2*N, 2*N))

DtauTilde = np.kron(D11,E11) + np.kron(D12,E12) + np.kron(D21,E21) + np.kron(D22,E22) 


def plot_L(ax, matrix, title, epsilon=1e-5):
    colors = np.zeros_like(matrix, dtype=np.dtype((float, 4)))  # RGBA

    for i in range(2*N):
        for j in range(2*N):
            value = matrix[i, j]
            if abs(value) < epsilon:
                colors[i, j] = [1, 1, 1, 1]  # Blanc pour les valeurs proches de zéro
            elif value > 0:
                colors[i, j] = [0, 0, 1, 1]  # Bleu pour les valeurs positives
            else:
                colors[i, j] = [1, 0, 0, 1]  # Rouge pour les valeurs négatives

    ax.imshow(colors, interpolation='none')
    # Contour noir
    ax.contour(colors[:, :, 0], levels=[2], colors='black', linewidths=2)
    
    # Grille
    ax.set_xticks(np.arange(0, 2*N, 1))
    ax.set_yticks(np.arange(0, 2*N, 1))
    ax.set_xticks(np.arange(-0.5, 2*N, 1), minor=True)
    ax.set_yticks(np.arange(-0.5, 2*N, 1), minor=True)
    ax.grid(which="minor", color="black", linestyle='-', linewidth=0.5)
    ax.tick_params(which="major", size=0)  # Désactiver les ticks majeurs

    
    ax.set_xticklabels([])  # Désactiver les labels des axes x
    ax.set_yticklabels([])  # Désactiver les labels des axes y

    ax.set_title(title)

# Création de sous-graphiques
fig, axs = plt.subplots(1, 2, figsize=(8, 8))


# Affichage des matrices
plot_L(axs[0], Dtau, r'$R$')
plot_L(axs[1], DtauTilde, r'$\tilde{R}$')

def plot_Lij(ax, matrix, title, epsilon=1e-5):
    colors = np.zeros_like(matrix, dtype=np.dtype((float, 4)))  # RGBA

    for i in range(N):
        for j in range(N):
            value = matrix[i, j]
            if abs(value) < epsilon:
                colors[i, j] = [1, 1, 1, 1]  # Blanc pour les valeurs proches de zéro
            elif value > 0:
                colors[i, j] = [0, 0, 1, 1]  # Bleu pour les valeurs positives
            else:
                colors[i, j] = [1, 0, 0, 1]  # Rouge pour les valeurs négatives

    ax.imshow(colors, interpolation='none')
    # Contour noir
    ax.contour(colors[:, :, 0], levels=[2], colors='black', linewidths=2)
    
    # Grille
    ax.set_xticks(np.arange(0, N, 1))
    ax.set_yticks(np.arange(0, N, 1))
    ax.set_xticks(np.arange(-0.5, N, 1), minor=True)
    ax.set_yticks(np.arange(-0.5, N, 1), minor=True)
    ax.grid(which="minor", color="black", linestyle='-', linewidth=0.5)
    ax.tick_params(which="major", size=0)  # Désactiver les ticks majeurs

    
    ax.set_xticklabels([])  # Désactiver les labels des axes x
    ax.set_yticklabels([])  # Désactiver les labels des axes y

    ax.set_title(title)

# Création de sous-graphiques
fig, axs = plt.subplots(1, 4, figsize=(12, 12))


# Affichage des matrices
plot_Lij(axs[0], D11, r'$R^{1,1}$')
plot_Lij(axs[2], D21, r'$R^{2,1}$')
plot_Lij(axs[1], D12, r'$R^{1,2}$')
plot_Lij(axs[3], D22, r'$R^{2,2}$')
```

C'est gagné ! Avec cette méthode, on arrive à avoir une matrice à bande, même pour un système d'équation d'ordre supérieur à un !

# Résumé de cette partie

### Principe numérique Global de la méthode tau

La méthode tau offre une implémentation élégante des conditions aux limites. En pratique, pour une équation d'ordre un avec une unique condition, l'application de la méthode tau revient à remplacer l'une des équations de la méthode de Galerkin par l'équation définie par la condition aux limites.

### Méthode de recombinaison de Dirichlet

Afin de préserver la structure creuse de la matrice tout en intégrant les conditions aux limites, nous avons la possibilité de jouer sur la base d'essai. En réalisant des combinaisons linéaires de la base initiale, il devient envisageable d'optimiser le nombre de zéros dans l'équation des conditions aux limites. La méthode de eecombinaison de Dirichlet se révèle particulièrement adaptée dans le contexte des polynômes de Chebychev.

### Systèmes d'Équations au-delà du Premier Ordre

Pour résoudre une équation d'ordre $N$, il est efficace de reformuler le problème en $N$ équations d'ordre un. L'application de la méthode tau s'effectue de manière modulaire, traitant chaque bloc individuellement. Cette approche conduit à une matrice composée de blocs à bande. En présence de blocs non nuls en dehors de la diagonale principale, il est avantageux d'utiliser le renversement du produit de Kronecker. Cela implique la décomposition de la matrice en un produit de Kronecker entre des matrices élémentaires et les blocs respectifs de la matrice. Le renversement de ce produit permet que la mtatrice résultante soit globalement à bande.
