# Principe numérique de la méthode tau

Le principe de la méthode tau est une extension de la méthode de Galerkin. Pour illustrer ce principe, considérons le cas où une seule condition aux limites de la forme $u(x_{CL}) = u_{CL}$ est présente, avec $x_{CL}$ représentant un point de l'espace.


## Ajout du terme tau et application de la méthode de Galerkin

En introduisant un terme tau dans l'équation de départ en raison d'une unique condition aux limites, nous obtenons :

$$ Lu(x) + \tau P(x) = f(x). $$

En appliquant la méthode de Galerkin (utilisant les mêmes notations que dans la première partie de ce projet), nous obtenons l'équation suivante pour tout $m \in \{0, 1, ..., N-1\}$ :

$$
\forall m \in \{0,1,...,N-1\}, \quad \sum_{n=0}^{N-1} u_n \langle\psi_m(x)|L|\phi_n(x)\rangle_2 + \tau \langle\psi_m(x)|P(x)\rangle_2 = \sum_{n=0}^{N-1} f_n \langle\psi_m(x)|\phi_n(x)\rangle_2.
$$

Cela nous donne un système avec $N+1$ inconnues ${u_0, ..., u_{N-1}, \tau}$ et $N$ équations.

En utilisant la liberté offerte par le paramètre $\tau$, nous pouvons ajouter une équation supplémentaire pour fermer le système. Pour cela, nous utiliserons la condition aux limites, ajoutant l'équation :

$$
\sum_{n=0}^{N-1} u_n |\phi_n(x_{CL})\rangle = u_{CL}.
$$

Ainsi, le système a désormais $N+1$ équations pour $N+1$ inconnues ${u_0, ..., u_{N-1}, \tau}$. Ce système peut être formulé de manière matricielle :


$$\begin{bmatrix}
\langle\psi_0|L|\phi_0\rangle_2 & \ldots & \langle\psi_0|L|\phi_{N-1}\rangle_2 & | & \langle\psi_0|P\rangle_2 \\
\vdots & \ddots & \vdots & | & \vdots \\
\langle\psi_{N-1}|L|\phi_0\rangle_2 & \ldots & \langle\psi_{N-1}|L|\phi_{N-1}\rangle_2 & | & \langle\psi_{N-1}|P\rangle_2 \\
\hline
\phi_{0}(x_{CL}) & \ldots & \phi_{N-1}(x_{CL}) & | & 0 \\
\end{bmatrix}
\begin{bmatrix}
U_0 \\
\vdots \\
U_{N-1} \\
\tau \\
\end{bmatrix}
= \begin{bmatrix}
\langle\psi_0|\phi_0\rangle_2 & \ldots & \langle\psi_0|\phi_{N-1}\rangle_2 & | & 0 \\
\vdots & \ddots & \vdots & | & \vdots \\
\langle\psi_{N-1}|\phi_0\rangle_2 & \ldots & \langle\psi_{N-1}|\phi_{N-1}\rangle_2 & | & 0 \\
0 & \ldots & 0 & | & 1 \\
\end{bmatrix}
\begin{bmatrix}
f_0 \\
\vdots \\
f_{N-1} \\
u_{CL} \\
\end{bmatrix}.
$$

## Choix du polynôme $P$

Comme expliqué dans la première partie de ce projet, il est numériquement avantageux de maximiser le nombre de zéros dans une matrice. Un choix efficace pour $P$ est un polynôme de la famille des $\phi_n$, par exemple :

$$ P = \psi_{N-1}(x). $$

Ainsi, avec $\psi_{n}(x)$ comme base orthonormale pour son produit scalaire, la formulation devient :


$$\begin{bmatrix}
\langle\psi_0|L|\phi_0\rangle_2 & \ldots & \langle\psi_0|L|\phi_{N-1}\rangle_2 & | & 0 \\
\vdots & \ddots & \vdots & | & \vdots \\
\langle\psi_{N-2}|L|\phi_0\rangle_2 & \ldots & \langle\psi_{N-2}|L|\phi_{N-2}\rangle_2 & | & 0\\
\langle\psi_{N-1}|L|\phi_0\rangle_2 & \ldots & \langle\psi_{N-1}|L|\phi_{N-1}\rangle_2 & | & 1\\
\hline
\phi_{0}(x_{CL}) & \ldots & \phi_{N-1}(x_{CL}) & | & 0 \\
\end{bmatrix}
\begin{bmatrix}
U_0 \\
\vdots \\
U_{N-1} \\
\tau \\
\end{bmatrix}= \begin{bmatrix}
\langle\psi_0|\phi_0\rangle_2 & \ldots & \langle\psi_0|\phi_{N-1}\rangle_2 & | & 0 \\
\vdots & \ddots & \vdots & | & \vdots \\
\langle\psi_{N-1}|\phi_0\rangle_2 & \ldots & \langle\psi_{N-1}|\phi_{N-1}\rangle_2 & | & 0 \\
\hline
0 & \ldots & 0 & | & 1 \\
\end{bmatrix}
\begin{bmatrix}
f_0 \\
\vdots \\
f_{N-1} \\
u_{CL} \\
\end{bmatrix}.
$$

En effectuant une permutation des deux dernières lignes, on obtient alors :


$$\begin{bmatrix}
\langle\psi_0|L|\phi_0\rangle_2 & \ldots & \langle\psi_0|L|\phi_{N-1}\rangle_2 & | & 0 \\
\vdots & \ddots & \vdots & | & \vdots \\
\langle\psi_{N-2}|L|\phi_0\rangle_2 & \ldots & \langle\psi_{N-2}|L|\phi_{N-2}\rangle_2 & | & 0\\
\phi_{0}(x_{CL}) & \ldots & \phi_{N-1}(x_{CL}) & | & 0 \\
\hline
\langle\psi_{N-1}|L|\phi_0\rangle_2 & \ldots & \langle\psi_{N-1}|L|\phi_{N-1}\rangle_2 & | & 1\\
\end{bmatrix}
\begin{bmatrix}
U_0 \\
\vdots \\
U_{N-1} \\
\tau \\
\end{bmatrix}= \begin{bmatrix}
\langle\psi_0|\phi_0\rangle_2 & \ldots & \langle\psi_0|\phi_{N-1}\rangle_2 & | & 0 \\
\vdots & \ddots & \vdots & | & \vdots \\
\langle\psi_{N-2}|\phi_0\rangle_2 & \ldots & \langle\psi_{N-2}|\phi_{N-1}\rangle_2 & | & 0 \\
0 & \ldots & 0 & | & 1 \\
\hline
\langle\psi_{N-1}|\phi_0\rangle_2 & \ldots & \langle\psi_{N-1}|\phi_{N-1}\rangle_2 & | & 0 \\
\end{bmatrix}
\begin{bmatrix}
f_0 \\
\vdots \\
f_{N-2} \\
f_{N-1} \\
u_{CL} \\
\end{bmatrix}.
$$

L'objectif étant de déterminer $u$, la valeur de $\tau$ ne nous intéresse pas. Nous pouvons donc éliminer son expression et l'équation qui le régit dans notre système. On obtient finalement :


$$\begin{bmatrix}
\langle\psi_0|L|\phi_0\rangle_2 & \ldots & \langle\psi_0|L|\phi_{N-1}\rangle_2  \\
\vdots & \ddots & \vdots  \\
\langle\psi_{N-2}|L|\phi_0\rangle_2 & \ldots & \langle\psi_{N-2}|L|\phi_{N-2}\rangle_2 \\
\phi_{0}(x_{CL}) & \ldots & \phi_{N-1}(x_{CL})  \\
\end{bmatrix}
\begin{bmatrix}
U_0 \\
\vdots \\
U_{N-1} \\
\end{bmatrix}= \begin{bmatrix}
\langle\psi_0|\phi_0\rangle_2 & \ldots & \langle\psi_0|\phi_{N-1}\rangle_2 & 0 \\
\vdots & \ddots & \vdots & \vdots \\
\langle\psi_{N-2}|\phi_0\rangle_2 & \ldots & \langle\psi_{N-2}|\phi_{N-1}\rangle_2 & 0 \\
0 & \ldots & 0 & 1 \\
\end{bmatrix}
\begin{bmatrix}
f_0 \\
\vdots \\
f_{N-2} \\
f_{N-1} \\
u_{CL} \\
\end{bmatrix}.
$$


## Écriture condensée du problème

On peut réécrire le problème de manière concise comme suit :

$$
D_{\tau}U=M_{\tau}F_{\tau}
$$

avec

- $D_{\tau}$ : la nouvelle matrice de dérivation contenant les conditions aux limites,
  $ D_{\tau_{ij}} =
\begin{cases}
\langle\psi_i(x)|L|\phi_j(x)\rangle_2 & \text{pour } i \neq N-1 \\
\phi_{j}(x_{CL}) & \text{pour } i = N-1
\end{cases} $

- $U$ : le vecteur des coefficients de $u$ dans la base, $U_{i}=u_i$

- $F_{\tau}$ : le vecteur des coefficients de $f$ dans la base et la condition aux limites,
  $ F_{\tau_{i}} =
\begin{cases}
f_i & \text{pour } i \neq N \\
u_{CL} & \text{pour } i = N
\end{cases} $

- $M_{\tau_{ij}}$ : la matrice de changement de base complétée,
  $ M_{ij} =
\begin{cases}
\langle\psi_i(x)|\phi_j(x)\rangle_2 & \text{pour } i \neq N-1  \text{et } j \neq N \\
\delta_{N-1,j} & \text{pour } i = N-1 \\
\delta_{i,N} & \text{pour } j = N
\end{cases} $

On retrouve ainsi un problème d'inversion de matrice similaire à celui de départ.





***
**Remarque : LHS, simplicité de la méthode tau**  

En termes de modification du terme de gauche (LHS), cette méthode se résume à la suppression de la dernière ligne, remplacée par la condition aux limites !
***

***
**Remarque : RHS, attention au terme ajouté** 

En revanche, pour le terme de droite (RHS), il est essentiel de le compléter. Si le LHS reste un produit d'une matrice $N\times N$ avec un vecteur de taille $N$, le RHS implique désormais le produit d'une matrice $N\times (N+1)$ avec un vecteur de taille $(N+1)$.
***

***
**Remarque : permutation des lignes**  

L'ordre dans lequel nous avons écrit les équations dans la matrice de dérivation est arbitraire. Il est tout à fait possible de permuter les lignes, mais il faut faire attention à également permuter les lignes de la matrice de changement, de base complétée.
***

## Application à un exemple 

Revenons à l'exemple jouet que nous avons utilisé dans la première partie de ce projet :
- $L=\frac{\partial}{\partial x}$,
- $|\phi_n(x)\rangle = T_n(x)$ avec son produit scalaire $ \langle ., . \rangle_1$,
- $|\psi_n(x)\rangle = U_n(x)$ avec son produit scalaire $ \langle ., . \rangle_2$.

Complétons-le avec la condition initiale arbitraire $u(1) = 0$ (par exemple une condition au bord fixe) et ajoutons-lui un terme tau avec $P=U_{N-1}(x)$, menant au système :

$$ \frac{\partial u}{\partial x} + \tau U_{N-1}(x)= f(x), \quad u(0) = 0.$$

Nous avons alors :

$$
D_{\tau}U=M_{\tau}F_{\tau}
$$

avec

- $ D_{\tau_{ij}} =
\begin{cases}
\langle U_n(x)|L|T_m(x)\rangle_2 = n\delta_{n,m-1}\langle U_n(x)|\ U_m(x)\rangle_2 & \text{pour } i \neq N-1 & \text{car } \frac{\partial T_n}{\partial x} = nU_{n-1}(x)\\
T_{j}(1) = 1 & \text{pour } i = N-1 & \text{car } T_n(1) = \cos(0) =1
\end{cases} $

- $U=u_i$

- $ F_{\tau_{i}} =
\begin{cases}
f_i & \text{pour } i \neq N \\
0 & \text{pour } i = N
\end{cases} $

- $ M_{ij} =
\begin{cases}
\langle U_i(x)|\ T_j(x)\rangle_2 & \text{pour } i \neq N-1  \text{ et } j \neq N \\
\delta_{N-1,j} & \text{pour } i = N-1 \\
\delta_{i,N} & \text{pour } j = N
\end{cases} $

La ligne de code ci-dessous permet de représenter graphiquement cet exemple.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Définition des paramètres
N = 15

# Définition des polynômes de Chebysev du premier ordre (T_n)
def Tn(n, x):
    return np.cos(n * np.arccos(x))

# Définition des polynômes de Chebysev du second ordre (U_n)
def Un(n, x):
    numerator = np.sin((n + 1) * np.arccos(x))
    denominator = np.sin(np.arccos(x))
    return numerator / denominator

# Définition des produits scalaires
def inner_product_0(f, g):
    integrand = lambda x: f(x) * g(x) / np.sqrt(1 - x**2)
    result, _ = quad(integrand, -1, 1)
    return result

def inner_product_2(f, g):
    integrand = lambda x: f(x) * g(x) * np.sqrt(1 - x**2)
    result, _ = quad(integrand, -1, 1)
    return result

# Calcul des dérivées des polynômes de Chebysev du premier ordre
def dTn_dx(n, x):
    return n * np.sin(n * np.arccos(x)) / np.sqrt(1 - x**2)

# Calcul des coefficents de la décomposition
Dtau = np.zeros((N, N))
Mtau = np.zeros((N, N+1))


for i in range(N-1):
    for j in range(N):
        f_Ti = lambda x: Tn(i, x)
        f_Tj = lambda x: Tn(j, x)
        f_Ui = lambda x: Un(i, x)
        f_Uj = lambda x: Un(j, x)
        f_dT = lambda x: dTn_dx(j, x)

        Dtau[i, j] = inner_product_2(f_Ui, f_dT)
        Mtau[i, j] = inner_product_2(f_Ui, f_Tj)

# Ajout de la condition initiale par la méthode tau
for i in range(N):
    Dtau[N-1, i] = 1
Mtau[N-1, N] = 1

# Fonction pour afficher la matrice avec des carrés noires pour les valeurs > epsilon et blanc sinon
def plot_matrix(ax, matrix, title, epsilon=1e-5):
    ax.imshow(np.abs(matrix) > epsilon, cmap='binary', interpolation='none')

    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            ax.text(j, i, f'{matrix[i, j]:.1f}', color='red',
                     ha='center', va='center', fontsize=8)

    ax.set_title(title)

# Création de sous-graphiques
fig, axs = plt.subplots(1, 2, figsize=(12, 12))


# Affichage des matrices
plot_matrix(axs[0], Dtau, r'$D_{\tau_{ij}}$')
plot_matrix(axs[1], Mtau, r'$M_{\tau_{ij}}$')
```

Par cet exemple, il devient évident que la méthode tau se révèle efficace pour l'implémentation des conditions aux limites. Néanmoins, elle introduit un nouveau défi : la matrice inverse n'est plus diagonale ni en bande.

> Une solution astucieuse pour restaurer une matrice diagonale ou en bande consiste à apporter de légères modifications à la base initiale. Cette approche permet de conserver la sparsité de la matrice de dérivation initiale tout en rendant également creuse la contribution due aux conditions aux limites. Nous appliquerons l'astuce de recombinaison de Dirichlet pour résoudre ce problème.

# Méthode de recombinaison de Dirichlet 

## Conditions de Dirichlet et de Neumann

Les conditions de Dirichlet et de Neumann sont des types de conditions aux limites fréquemment employées dans la résolution d'équations aux dérivées partielles, en particulier dans le cadre de la physique. Voici une synthèse de chacune de ces conditions :

1. **Condition de Dirichlet :**
   - La condition de Dirichlet spécifie la valeur d'une fonction à la frontière d'un domaine.
   - Un exemple concret serait, dans le contexte de l'équation modélisant la distribution thermique dans un objet, la fixation de la température à la surface de cet objet.

2. **Condition de Neumann :**
   - La condition de Neumann spécifie le taux de variation d'une fonction à la frontière d'un domaine.
   - Plutôt que de fixer la valeur de la fonction à la frontière, la condition de Neumann fixe la dérivée normale de la fonction à cet endroit.
   - On peut interpréter cela comme un flux ou un taux de transfert à travers la frontière du domaine.
   - Par exemple, dans le contexte de la conduction thermique, la condition de Neumann pourrait définir le flux de chaleur à travers la frontière d'un objet.

En résumé, la condition de Dirichlet fixe les valeurs de la fonction à la frontière, tandis que la condition de Neumann fixe les dérivées normales à la frontière.

Il est important de noter que lors de l'adimensionnement, les frontières sont souvent représentées par des expressions du type $x=1$ ou $x=-1$. C'est pourquoi, dans le cas des polynômes de Chebysev, nous optons généralement pour le préconditionnement de Dirichlet.

## Préconditionnement de Dirichlet

On définit une nouvelle base d'essaie (noté avant $\phi_n(x)$): 
$$
D_n(x) = 
\begin{cases}
T_n(x), & \text{pour } n = 0,1 \\
T_n(x) - T_{n-2}(x), & \text{pour } n > 1
\end{cases}
\qquad (12)
$$

On a alors pour cette famille au niveau des bords du type $x=1$ ou $x=-1$ :
$$
D_n(\pm 1) = 
\begin{cases}
(\pm 1)^n, & \text{pour } n = 0,1 \\
0, & \text{pour } n > 1
\end{cases}
$$

Dans cette base, les conditions aux limites de Dirichlet ne concernent que les deux premiers coefficients d'expansion. Cette technique est connue sous le nom de "préconditionnement de Dirichlet" ou "recombinaison de base".

Ainsi, si on veut retrouver une matrice diagonale ou en bande, il est important de permuter les lignes de sorte que les deux coefficients soient en haut à gauche, et non pas en haut à droite. 

Visuellement on a initalement :
$$
\begin{bmatrix}
\langle\psi_0|L|D_0\rangle_2 & \ldots & \langle\psi_0|L|D_{N-1}\rangle_2  \\
\vdots & \ddots & \vdots  \\
\langle\psi_{N-2}|L|D_0\rangle_2 & \ldots & \langle\psi_{N-2}|L|D_{N-2}\rangle_2 \\
D_{0}(x_{CL}) & \ldots & D_{N-1}(x_{CL})  \\
\end{bmatrix}
\begin{bmatrix}
U_0 \\
\vdots \\
U_{N-1} \\
\end{bmatrix}= \begin{bmatrix}
\langle\psi_0|D_0\rangle_2 & \ldots & \langle\psi_0|D_{N-1}\rangle_2 & 0 \\
\vdots & \ddots & \vdots & \vdots \\
\langle\psi_{N-2}|D_0\rangle_2 & \ldots & \langle\psi_{N-2}|D_{N-1}\rangle_2 & 0 \\
0 & \ldots & 0 & 1 \\
\end{bmatrix}
\begin{bmatrix}
f_0 \\
\vdots \\
f_{N-2} \\
f_{N-1} \\
u_{CL} \\
\end{bmatrix}.
$$

On évalue $D_n$ en $x=\pm 1$ :

$$
\begin{bmatrix}
\langle\psi_0|L|D_0\rangle_2 & & \ldots & & \langle\psi_0|L|D_{N-1}\rangle_2  \\
\vdots & & \ddots & & \vdots  \\
\langle\psi_{N-2}|L|D_0\rangle_2 & &  \ldots & & \langle\psi_{N-2}|L|D_{N-2}\rangle_2 \\
(\pm 1)^n & (\pm 1)^n & 0 & \ldots & 0  \\
\end{bmatrix}
\begin{bmatrix}
U_0 \\
\vdots \\
U_{N-1} \\
\end{bmatrix}= \begin{bmatrix}
\langle\psi_0|D_0\rangle_2 & \ldots & \langle\psi_0|D_{N-1}\rangle_2 & 0 \\
\vdots & \ddots & \vdots & \vdots \\
\langle\psi_{N-2}|D_0\rangle_2 & \ldots & \langle\psi_{N-2}|D_{N-1}\rangle_2 & 0 \\
0 & \ldots & 0 & 1 \\
\end{bmatrix}
\begin{bmatrix}
f_0 \\
\vdots \\
f_{N-2} \\
f_{N-1} \\
u_{CL} \\
\end{bmatrix}.
$$

Puis on permute les lignes dans l'objectif d'avoir une matrice en bande.

$$
\begin{bmatrix}
(\pm 1)^n & (\pm 1)^n & 0 & \ldots & 0  \\
\langle\psi_0|L|D_0\rangle_2 & & \ldots & & \langle\psi_0|L|D_{N-1}\rangle_2  \\
\vdots & & \ddots & & \vdots  \\
\langle\psi_{N-2}|L|D_0\rangle_2 & &  \ldots & & \langle\psi_{N-2}|L|D_{N-2}\rangle_2 \\
\end{bmatrix}
\begin{bmatrix}
U_0 \\
\vdots \\
U_{N-1} \\
\end{bmatrix}= \begin{bmatrix}
0 & \ldots & 0 & 1 \\
\langle\psi_0|D_0\rangle_2 & \ldots & \langle\psi_0|D_{N-1}\rangle_2 & 0 \\
\vdots & \ddots & \vdots & \vdots \\
\langle\psi_{N-2}|D_0\rangle_2 & \ldots & \langle\psi_{N-2}|D_{N-1}\rangle_2 & 0 \\
\end{bmatrix}
\begin{bmatrix}
f_0 \\
\vdots \\
f_{N-2} \\
f_{N-1} \\
u_{CL} \\
\end{bmatrix},
$$

***
**Remarque : Diagonalité de la matrice de changement de base**

Il est avantageux que la matrice de changement de base complétée contienne un maximum de zéros pour faciliter le produit matriciel. Cependant, l'aspect crucial réside dans le fait que la matrice de dérivation avec les conditions limites soit aussi proche que possible d'une matrice diagonale. En effet, le calcul de l'inversion de matrice est bien plus coûteux que celui du produit matriciel. **On s'intéressera donc principalement dans la suite à l'optimisation de la matrice $D_{\tau}$.**
***


## Application à un exemple 

On reprend notre exemple en changeant $|\phi_n(x)\rangle$ :
$$|\phi_n(x)\rangle = D_n(x).$$


Nous avons alors :

$$
D_{\tau}U=M_{\tau}F_{\tau}
$$

avec

- $ D_{\tau_{ij}} =
\begin{cases}
\langle U_n(x)|L|D_m(x)\rangle_2 & \text{pour } i \neq 0 \\
D_{j}(1) = 1 & \text{pour } i = 0 \text{ et } j = 0 \text{ ou } 1\\
D_{j}(1) = 0 & \text{pour } i = 0 \text{ et } j >1
\end{cases} $


La ligne de code ci-dessous permet de représenter graphiquement cet exemple.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Définition des paramètres
N = 15

# Définition des polynômes de Chebysev du premier ordre (T_n)
def Tn(n, x):
    return np.cos(n * np.arccos(x))

# Définition des polynômes de Chebysev du second ordre (U_n)
def Un(n, x):
    numerator = np.sin((n + 1) * np.arccos(x))
    denominator = np.sin(np.arccos(x))
    return numerator / denominator

# Définition des produits scalaires
def inner_product_0(f, g):
    integrand = lambda x: f(x) * g(x) / np.sqrt(1 - x**2)
    result, _ = quad(integrand, -1, 1)
    return result

def inner_product_2(f, g):
    integrand = lambda x: f(x) * g(x) * np.sqrt(1 - x**2)
    result, _ = quad(integrand, -1, 1)
    return result

# Calcul des dérivées des polynômes de Chebysev du premier ordre
def dTn_dx(n, x):
    return n * np.sin(n * np.arccos(x)) / np.sqrt(1 - x**2)

# Calcul des polynomes de Dirichlet
def Dn(n, x):
    if n == 0 or n == 1:
        return Tn(n, x)
    elif n > 1:
        return Tn(n, x) - Tn(n-2, x)
        
# Et de leurs dérivés 
def dDn_dx(n, x):
    if n == 0 or n == 1:
        return dTn_dx(n, x)
    elif n > 1:
        return dTn_dx(n, x) - dTn_dx(n-2, x)

# Calcul des coefficents de la décomposition
Dtau = np.zeros((N, N))
Mtau = np.zeros((N, N+1))


for i in range(2,N):
    for j in range(N):
        f_Ui = lambda x: Un(i, x)
        f_dD = lambda x: dDn_dx(j, x)

        Dtau[i, j] = inner_product_2(f_Ui, f_dD)

# Ajout de la condition initiale par la méthode tau
for i in range(2):
    Dtau[0, i] = 1

# Fonction pour afficher la matrice avec des carrés noires pour les valeurs > epsilon et blanc sinon
def plot_matrix(ax, matrix, title, epsilon=1e-5):
    ax.imshow(np.abs(matrix) > epsilon, cmap='binary', interpolation='none')

    for i in range(matrix.shape[0]):
        for j in range(matrix.shape[1]):
            ax.text(j, i, f'{matrix[i, j]:.1f}', color='red',
                     ha='center', va='center', fontsize=8)

    ax.set_title(title)

# Création de sous-graphiques
fig = plt.figure(figsize=(8, 8))
axs = fig.add_subplot(1, 1, 1)

# Affichage des matrices
plot_matrix(axs, Dtau, r'$D_{\tau_{ij}}$')
```

***
**Remarque : Matrice bande**

Il est essentiel de noter que l'obtention de matrices diagonales n'est réalisable que dans des cas très simples. Dans la suite, nous nous contenterons d'obtenir des matrices bande. Cela signifie qu'il s'agit de matrices creuses, dont les coefficients non nuls sont limités à une bande diagonale, englobant la diagonale principale et éventuellement une ou plusieurs diagonales de chaque côté. Ces matrices, après les matrices diagonales, sont les matrices les plus simples à inverser et à stocker.

L'objectif de cette section est d'explorer l'implémentation numérique de la méthode tau, en se basant sur l'exemple de l'équation de Poisson à une dimension :

$$ \frac{\partial^2 u}{\partial x^2} = f(x), \quad u(-1) = 0, \quad \frac{\partial u}{\partial x}(1) = 0. $$

Pour la méthode de Galerkin, nous utiliserons les familles de polynômes de Chebyshev (définies dans la première partie de ce projet) :
- $|\phi_n(x)\rangle = T_n(x)$ 
- $|\psi_n(x)\rangle = U_n(x)$

Avec leurs produits scalaires associés notés $ \langle ., . \rangle_1$ et $ \langle ., . \rangle_2$.


# A. Méthode de Galerkin pour tout les ordres de dérivées

Dans la première partie de ce projet, en appliquant la méthode de Galerkin à l'équation $Lu(x) = f(x)$, nous avions abouti à l'équation matricielle suivante :

$$
DU=MF,
$$
avec 
- $D$ la matrice dérivartion défini par : $D_{ij} = \langle\psi_i(x)|L|\phi_j(x)\rangle_2$
- $U$ le vecteur des coefficients de $u$ dans la base : $U_{i}=u_i$
- $F$ le vecteur des coefficients de $f$ dans la base : $F_{i}=f_i$
- $B_1$ la matrice de la base d'essaie : $B_{1,ij} = \langle\phi_i(x)|\phi_j(x)\rangle_1$
- $B_2$ la matrice de la base de test: $B_{2,ij} = \langle\psi_i(x)|\psi_j(x)\rangle_2$
- $M$ la matrice de changement de base : $M_{ij} = \langle\psi_i(x)|\phi_j(x)\rangle_2$

Pour une équation comme celle de l'exemple d'ordre deux, on pourrait appliquer la méthode tel quel. Seulement le nombre de termes tau et le nombre de conditions aux limites correspondront au nombre total de dérivées dans le système. Il est donc plus facile de les compter en convertissant le système en une forme du premier ordre, 





Leur $L$ :

$$ L = \begin{pmatrix}
  0 & \partial_x \\
  \partial_x & -1 \\
\end{pmatrix} $$

## T-to-T equation blocks

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Définition des paramètres
N = 24 #

# Définition des polynômes de Chebyshev 
def Tn(n, x):
    """
    Calcule la n-ième fonction de base des polynômes de Chebyshev 
    
    Paramètres :
        - x : Valeurs de la variable indépendante
        - n : Indice de la fonction de base
    
    Retourne :
        - Valeurs de la n-ième fonction de base des polynômes de Chebyshev  évaluées en x
    """
    return np.cos(n * np.arccos(x))

# Définition du produit scalaire
def inner_product(f, g):
    integrand = lambda x: f(x) * g(x) / np.sqrt(1 - x**2)
    result, _ = quad(integrand, -1, 1)
    return result


# Calcul des dérivées des polynômes de Chebysev 
def dTn_dx(n, x):
    return n * np.sin(n * np.arccos(x)) / np.sqrt(1 - x**2)

# Défintion de L 
L = np.zeros((2*N, 2*N))

# Calcul des coefficents de la matrice de dérivation et de la base
D = np.zeros((N, N))
B = np.zeros((N, N))

for i in range(N):
    for j in range(N):
        f_Ti = lambda x: Tn(i, x)
        f_Tj = lambda x: Tn(j, x)
        f_dT = lambda x: dTn_dx(j, x)

        D[i, j] = inner_product(f_Ti, f_dT)
        B[i, j] = inner_product(f_Ti, f_Tj)

L[N:,0:N] = D
L[0:N,N:] = D
L[N:,N:] = -B

# Fonction pour afficher la matrice avec des carrés noires pour les valeurs > epsilon et blanc sinon
def plot_matrix(ax, matrix, title, epsilon=1e-5):
    ax.imshow(np.abs(matrix) > epsilon, cmap='binary', interpolation='none')

    for i in range(N):
        for j in range(N):
            ax.text(j, i, f'{matrix[i, j]:.1f}', color='red',
                     ha='center', va='center', fontsize=8)

    ax.set_title(title)

def plot_L(ax, matrix, title, epsilon=1e-5):
    colors = np.zeros_like(matrix, dtype=np.dtype((float, 4)))  # RGBA

    for i in range(2*N):
        for j in range(2*N):
            value = matrix[i, j]
            if abs(value) < epsilon:
                colors[i, j] = [1, 1, 1, 1]  # Blanc pour les valeurs proches de zéro
            elif value > 0:
                colors[i, j] = [0, 0, 1, 1]  # Bleu pour les valeurs positives
            else:
                colors[i, j] = [1, 0, 0, 1]  # Rouge pour les valeurs négatives

    ax.imshow(colors, interpolation='none')
    # Contour noir
    ax.contour(colors[:, :, 0], levels=[2], colors='black', linewidths=2)
    
    # Grille
    ax.set_xticks(np.arange(0, 2*N, 1))
    ax.set_yticks(np.arange(0, 2*N, 1))
    ax.set_xticks(np.arange(-0.5, 2*N, 1), minor=True)
    ax.set_yticks(np.arange(-0.5, 2*N, 1), minor=True)
    ax.grid(which="minor", color="black", linestyle='-', linewidth=0.5)
    ax.tick_params(which="major", size=0)  # Désactiver les ticks majeurs

    
    ax.set_xticklabels([])  # Désactiver les labels des axes x
    ax.set_yticklabels([])  # Désactiver les labels des axes y

    ax.set_title(title)
    


fig, axs = plt.subplots(1, 3, figsize=(12, 12))


# Affichage des matrices
plot_matrix(axs[0], B, r'$B_{ij} = \langle T_i(x)| T_j(x)\rangle$')
plot_matrix(axs[1], D, r'$D_{ij} = \langle T_i(x)|L| T_j(x)\rangle$')
plot_L(axs[2], L, r'$L$')
```

## Ajout des conditions aux limites

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Définition des paramètres
N = 24

# Définition des polynômes de Chebysev du premier ordre (T_n)
def Tn(n, x):
    return np.cos(n * np.arccos(x))

# Définition des polynômes de Chebysev du second ordre (U_n)
def Un(n, x):
    numerator = np.sin((n + 1) * np.arccos(x))
    denominator = np.sin(np.arccos(x))
    return numerator / denominator

# Définition des produits scalaires
def inner_product_1(f, g):
    integrand = lambda x: f(x) * g(x) / np.sqrt(1 - x**2)
    result, _ = quad(integrand, -1, 1)
    return result

def inner_product_2(f, g):
    integrand = lambda x: f(x) * g(x) * np.sqrt(1 - x**2)
    result, _ = quad(integrand, -1, 1)
    return result

# Calcul des dérivées des polynômes de Chebysev du premier ordre
def dTn_dx(n, x):
    return n * np.sin(n * np.arccos(x)) / np.sqrt(1 - x**2)

# Calcul des coefficents de la décomposition
B_1 = np.zeros((N, N))
B_2 = np.zeros((N, N))
D = np.zeros((N, N))
M = np.zeros((N, N))

# Défintion de L 
L = np.zeros((2*N, 2*N))

# Ajout du terme tau pour la conditon aux limites 
tau = 1/2


for i in range(N):
    for j in range(N):
        f_Ti = lambda x: Tn(i, x)
        f_Tj = lambda x: Tn(j, x)
        f_Ui = lambda x: Un(i, x)
        f_Uj = lambda x: Un(j, x)
        f_dT = lambda x: dTn_dx(j, x)

        D[i, j] = inner_product_2(f_Ui, f_dT)
        B_1[i, j] = inner_product_1(f_Ti, f_Tj)
        B_2[i, j] = inner_product_2(f_Ui, f_Uj)
        M[i, j] = inner_product_2(f_Ui, f_Tj)

L[N:,0:N] = D
L[0:N-1,N:] = D
L[N:,N:] = -M
# tau method 
L[N-1,N:] = B_2[N-1, :]


# Fonction pour afficher la matrice avec des carrés noires pour les valeurs > epsilon et blanc sinon
def plot_matrix(ax, matrix, title, epsilon=1e-5):
    ax.imshow(np.abs(matrix) > epsilon, cmap='binary', interpolation='none')

    for i in range(N):
        for j in range(N):
            ax.text(j, i, f'{matrix[i, j]:.1f}', color='red',
                     ha='center', va='center', fontsize=8)

    ax.set_title(title)

def plot_L(ax, matrix, title, epsilon=1e-5):
    colors = np.zeros_like(matrix, dtype=np.dtype((float, 4)))  # RGBA

    for i in range(2*N):
        for j in range(2*N):
            value = matrix[i, j]
            if abs(value) < epsilon:
                colors[i, j] = [1, 1, 1, 1]  # Blanc pour les valeurs proches de zéro
            elif value > 0:
                colors[i, j] = [0, 0, 1, 1]  # Bleu pour les valeurs positives
            else:
                colors[i, j] = [1, 0, 0, 1]  # Rouge pour les valeurs négatives

    ax.imshow(colors, interpolation='none')
    # Contour noir
    ax.contour(colors[:, :, 0], levels=[2], colors='black', linewidths=2)
    
    # Grille
    ax.set_xticks(np.arange(0, 2*N, 1))
    ax.set_yticks(np.arange(0, 2*N, 1))
    ax.set_xticks(np.arange(-0.5, 2*N, 1), minor=True)
    ax.set_yticks(np.arange(-0.5, 2*N, 1), minor=True)
    ax.grid(which="minor", color="black", linestyle='-', linewidth=0.5)
    ax.tick_params(which="major", size=0)  # Désactiver les ticks majeurs

    
    ax.set_xticklabels([])  # Désactiver les labels des axes x
    ax.set_yticklabels([])  # Désactiver les labels des axes y

    ax.set_title(title)

# Création de sous-graphiques
fig, axs = plt.subplots(1, 4, figsize=(12, 12))

# Utilisation de add_subplot pour figL
figL = plt.figure(figsize=(8, 8))
axsL = figL.add_subplot(1, 1, 1)



# Affichage des matrices
plot_matrix(axs[0], B_1, r'$B_{1,ij} = \langle T_i(x)|T_j(x)\rangle_1$')
plot_matrix(axs[1], B_2, r'$B_{2,ij} = \langle U_i(x)|U_j(x)\rangle_2$')
plot_matrix(axs[2], D, r'$D_{ij} = \langle U_i(x)|L| T_j(x)\rangle_2$')
plot_matrix(axs[3], M, r'$M_{ij} = \langle U_i(x)| T_j(x)\rangle_2$')
plot_L(axsL, L, r'$L$')
```

## Dirichlet Recombination 

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad

# Définition des paramètres
N = 10

# Définition des polynômes de Chebysev du premier ordre (T_n)
def Tn(n, x):
    return np.cos(n * np.arccos(x))

# Définition des polynômes de Chebysev du second ordre (U_n)
def Un(n, x):
    numerator = np.sin((n + 1) * np.arccos(x))
    denominator = np.sin(np.arccos(x))
    return numerator / denominator

# Définition des produits scalaires
def inner_product_1(f, g):
    integrand = lambda x: f(x) * g(x) / np.sqrt(1 - x**2)
    result, _ = quad(integrand, -1, 1)
    return result

def inner_product_2(f, g):
    integrand = lambda x: f(x) * g(x) * np.sqrt(1 - x**2)
    result, _ = quad(integrand, -1, 1)
    return result

# Calcul des dérivées des polynômes de Chebysev du premier ordre
def dTn_dx(n, x):
    return n * np.sin(n * np.arccos(x)) / np.sqrt(1 - x**2)


# Définition des polynôme de Dirichlet 
def Diri(n, x):
    if n>1 :
        return Tn(n, x) - Tn(n-2, x)
    else :
        return Tn(n, x)
         
def dDiri_x(n, x):
    if n>1 :
        return dTn_dx(n, x) - dTn_dx(n-2, x)
    else :
        return dTn_dx(n, x)

# Calcul des coefficents de la décomposition
B_1 = np.zeros((N, N))
B_2 = np.zeros((N, N))
D = np.zeros((N, N))
M = np.zeros((N, N))


# Défintion de L 
L = np.zeros((2*N, 2*N))

for i in range(N):
    for j in range(N):
        f_Ti = lambda x: Tn(i, x)
        f_Tj = lambda x: Tn(j, x)
        f_Ui = lambda x: Un(i, x)
        f_Uj = lambda x: Un(j, x)
        f_dT = lambda x: dTn_dx(j, x)
        f_Dirii = lambda x: Diri(i, x)
        f_Dirij = lambda x: Diri(j, x)
        f_dDiri = lambda x: dDiri_x(j, x)

        D[i, j] = inner_product_2(f_Ui, f_dDiri)
        B_1[i, j] = inner_product_1(f_Dirii, f_Dirij)
        B_2[i, j] = inner_product_2(f_Ui, f_Uj)
        M[i, j] = inner_product_2(f_Ui, f_Dirij)

L[N:,0:N] = D
L[0:N,N:] = D
L[N:,N:] = -M

# Fonction pour afficher la matrice avec des carrés noires pour les valeurs > epsilon et blanc sinon
def plot_matrix(ax, matrix, title, epsilon=1e-1):
    ax.imshow(np.abs(matrix) > epsilon, cmap='binary', interpolation='none')

    for i in range(N):
        for j in range(N):
            ax.text(j, i, f'{matrix[i, j]:.1f}', color='red',
                     ha='center', va='center', fontsize=8)

    ax.set_title(title)

def plot_L(ax, matrix, title, epsilon=1e-5):
    colors = np.zeros_like(matrix, dtype=np.dtype((float, 4)))  # RGBA

    for i in range(2*N):
        for j in range(2*N):
            value = matrix[i, j]
            if abs(value) < epsilon:
                colors[i, j] = [1, 1, 1, 1]  # Blanc pour les valeurs proches de zéro
            elif value > 0:
                colors[i, j] = [0, 0, 1, 1]  # Bleu pour les valeurs positives
            else:
                colors[i, j] = [1, 0, 0, 1]  # Rouge pour les valeurs négatives

    ax.imshow(colors, interpolation='none')
    # Contour noir
    ax.contour(colors[:, :, 0], levels=[2], colors='black', linewidths=2)
    
    # Grille
    ax.set_xticks(np.arange(0, 2*N, 1))
    ax.set_yticks(np.arange(0, 2*N, 1))
    ax.set_xticks(np.arange(-0.5, 2*N, 1), minor=True)
    ax.set_yticks(np.arange(-0.5, 2*N, 1), minor=True)
    ax.grid(which="minor", color="black", linestyle='-', linewidth=0.5)
    ax.tick_params(which="major", size=0)  # Désactiver les ticks majeurs

    
    ax.set_xticklabels([])  # Désactiver les labels des axes x
    ax.set_yticklabels([])  # Désactiver les labels des axes y

    ax.set_title(title)

# Création de sous-graphiques
fig, axs = plt.subplots(1, 4, figsize=(12, 12))

# Utilisation de add_subplot pour figL
figL = plt.figure(figsize=(8, 8))
axsL = figL.add_subplot(1, 1, 1)



# Affichage des matrices
plot_matrix(axs[0], B_1, r'$B_{1,ij} = \langle T_i(x)|T_j(x)\rangle_1$')
plot_matrix(axs[1], B_2, r'$B_{2,ij} = \langle U_i(x)|U_j(x)\rangle_2$')
plot_matrix(axs[2], D, r'$D_{ij} = \langle U_i(x)|L| T_j(x)\rangle_2$')
plot_matrix(axs[3], M, r'$M_{ij} = \langle U_i(x)| T_j(x)\rangle_2$')
plot_L(axsL, L, r'$L$')
```
